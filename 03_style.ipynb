```
import numpy as np
import matplotlib.pyplot as plt
```

```
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)

plt.plot(x, x_sin)
```

```
plt.style.use('dark_background')
plt.plot(x, x_sin)
```

```
plt.style.use('seaborn')
plt.plot(x, x_sin)
```
