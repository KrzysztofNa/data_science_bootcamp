```
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```
import seaborn as sns
sns.set()
```

```
fig = plt.figure()
fig.show()
```

```
plt.plot([0,1], [0,3], label = 'line', color='green')
plt.xlabel('oś X')
plt.ylabel('oś Y')
plt.title('Wykres')
_ = plt.legend()
```

```
plt.plot([0,1, 2], [0,3,2], label = 'line', color='green')
plt.xlabel('oś X')
plt.ylabel('oś Y')
plt.title('Wykres')
_ = plt.legend()
```

```
x = np.arange(-3,3, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)

plt.plot(x, x_sin)
plt.plot(x, x_cos)
```

```
plt.subplot(121)
plt.plot([0,1,2], [3,2,5])
plt.subplot(122)
plt.plot([0,1,2], [3,1,4], 'r--')
```

```
plt.subplot(131)
plt.plot([0,1,2], [3,2,5])
plt.subplot(132)
plt.plot([0,1,2], [3,1,4], 'r--')
plt.subplot(133)
plt.plot([0,1,2], [3,1,4], 'ro')
```
